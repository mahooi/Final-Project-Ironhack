```python
import pandas as pd
import time
import datetime as dt
import numpy as np
import json
import csv
import matplotlib as plot
import re
```

```python
df=pd.read_csv('../Data/BIG_BOI.csv')
```

```python
df
```

```python
baking= (df.drop(columns=['Unnamed: 0'])
                                                        .sort_values('date', ascending=False)
                                                        .reset_index(drop='true')
                                                        .drop_duplicates()
                                                        
              )                                                       
```

```python
new_date=[]
for day in baking['date'].to_list():
    ndate=re.split("\s", day, 1)
    new_date.append(ndate[0])
baking['date']=new_date
```

```python



baking.plot(x='date',y='id', color='green',linewidth=3, )
```
